```python
import networkx as nx
import pylab as plt

x=nx.DiGraph()

x.add_weighted_edges_from([('A','B',1),('C','A',1),('B','C',1),('D','A',1),('E','A',1),('C','D',1),('D','B',1),('D','E',1),('A','E',1)],('C','E','1'))
print (nx.pagerank(x))

#graph plot
nx.draw(x, with_labels=True)
plt.show()
```
